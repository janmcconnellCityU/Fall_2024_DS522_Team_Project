```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the necessary files
salaries_df = pd.read_csv(r'C:\Users\JanMc\Dropbox\OneDrive\Documents\GitHub\janmcconnellCityU\Fall2024\DS522\TeamProject\data\players\Salaries_1985to2018.csv')
players_df = pd.read_csv(r'C:\Users\JanMc\Dropbox\OneDrive\Documents\GitHub\janmcconnellCityU\Fall2024\DS522\TeamProject\data\players\players.csv')
shots_df = pd.read_csv(r'C:\Users\JanMc\Dropbox\OneDrive\Documents\GitHub\janmcconnellCityU\Fall2024\DS522\TeamProject\data\shots\NBA_2004_Shots.csv')

# Step 1: Merge salaries with the players data to get player names
salaries_with_names_df = pd.merge(salaries_df, players_df, left_on='player_id', right_on='_id', how='inner')

# Step 2: Merge the resulting data with shots data on player name
merged_df = pd.merge(shots_df, salaries_with_names_df, left_on='PLAYER_NAME', right_on='name', how='inner')

# Filter for the years 2004 to 2018
filtered_df = merged_df[(merged_df['season_start'] >= 2004) & (merged_df['season_start'] <= 2018)]

# Step 3: Filter for 'Made Shot' event type
made_shots_df = filtered_df[filtered_df['EVENT_TYPE'] == 'Made Shot']

# Step 4: Group by player to count the number of made shots per player
made_shots_per_player = made_shots_df.groupby('PLAYER_NAME')['EVENT_TYPE'].count().reset_index(name='made_shots')

# Step 5: Merge made shots per player with salary data
salary_made_shots_df = pd.merge(made_shots_per_player, salaries_with_names_df, left_on='PLAYER_NAME', right_on='name', how='inner')

# Step 6: Calculate correlation between salary and made shots
correlation = salary_made_shots_df['salary'].corr(salary_made_shots_df['made_shots'])
print(f'Correlation between salary and made shots: {correlation}')

# Step 7: Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='made_shots', data=salary_made_shots_df)
plt.title("Correlation between Salary and Made Shots (2004-2018)")
plt.xlabel("Salary (millions)")
plt.ylabel("Made Shots")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Merge salaries with the players data to get player names
salaries_with_names_df = pd.merge(salaries_df, players_df, left_on='player_id', right_on='_id', how='inner')

# Step 2: Merge the resulting data with shots data on player name and filter for years 2004-2018
merged_df = pd.merge(shots_df, salaries_with_names_df, left_on='PLAYER_NAME', right_on='name', how='inner')

# Filter for years 2004 to 2018
filtered_df = merged_df[(merged_df['season_start'] >= 2004) & (merged_df['season_start'] <= 2018)]

# Step 3: Filter for 'Made Shot' event type
made_shots_df = filtered_df[filtered_df['EVENT_TYPE'] == 'Made Shot']

# Step 4: Group by player and year to count the number of made shots per player per year
made_shots_per_player_year = made_shots_df.groupby(['PLAYER_NAME', 'season_start'])['EVENT_TYPE'].count().reset_index(name='made_shots')

# Step 5: Merge made shots per player with salary data
salary_made_shots_df = pd.merge(made_shots_per_player_year, salaries_with_names_df, left_on=['PLAYER_NAME', 'season_start'], right_on=['name', 'season_start'], how='inner')

# Step 6: Calculate correlation between salary and made shots for each year
correlation_by_year = salary_made_shots_df.groupby('season_start').apply(lambda x: x['salary'].corr(x['made_shots']))
print(f'Correlation between salary and made shots by year (2004-2018):\n{correlation_by_year}')

# Step 7: Visualize the relationship per year using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='made_shots', hue='season_start', data=salary_made_shots_df, palette='coolwarm', legend='full')
plt.title("Correlation between Salary and Made Shots (2004-2018) by Year")
plt.xlabel("Salary (millions)")
plt.ylabel("Made Shots")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Sort the data by PLAYER_NAME and season_start to facilitate difference calculations
salary_made_shots_df = salary_made_shots_df.sort_values(by=['PLAYER_NAME', 'season_start'])

# Step 2: Calculate the year-over-year change in salary and made shots for each player
salary_made_shots_df['salary_increase'] = salary_made_shots_df.groupby('PLAYER_NAME')['salary'].diff()
salary_made_shots_df['shots_increase'] = salary_made_shots_df.groupby('PLAYER_NAME')['made_shots'].diff()

# Step 3: Drop any rows with NaN values resulting from the diff operation (first year per player will have NaN)
salary_made_shots_df = salary_made_shots_df.dropna(subset=['salary_increase', 'shots_increase'])

# Step 4: Calculate the correlation between salary increase and change in made shots
correlation_increase = salary_made_shots_df['salary_increase'].corr(salary_made_shots_df['shots_increase'])
print(f'Correlation between salary increase and made shots increase: {correlation_increase}')

# Step 5: Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary_increase', y='shots_increase', data=salary_made_shots_df)
plt.title("Correlation between Salary Increase and Made Shots Increase (2004-2018)")
plt.xlabel("Salary Increase (millions)")
plt.ylabel("Made Shots Increase")
plt.show()
```

```python
# Step 1: Identify the top 10 players by total salary over the years
top_10_players = salary_made_shots_df.groupby('PLAYER_NAME')['salary'].sum().nlargest(10).index

# Step 2: Filter the data to keep only the records for the top 10 players
top_10_df = salary_made_shots_df[salary_made_shots_df['PLAYER_NAME'].isin(top_10_players)]

# Step 3: Calculate the correlation between salary and made shots for the top 10 players
correlation_top_10 = top_10_df['salary'].corr(top_10_df['made_shots'])
print(f'Correlation between salary and made shots for top 10 players: {correlation_top_10}')

# Step 4: Visualize the relationship for the top 10 players
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='made_shots', hue='PLAYER_NAME', data=top_10_df, palette='tab10', legend='full')
plt.title("Correlation between Salary and Made Shots for Top 10 Players (2004-2018)")
plt.xlabel("Salary (millions)")
plt.ylabel("Made Shots")
plt.legend(title="Player")
plt.show()
```
